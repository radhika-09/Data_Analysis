```python
df.set_index("Date")
```

```python
plt.figure(figsize=(14,10))
df["Close"].plot()
plt.show()
```

```python
df_close=df["Close"]
plt.figure(figsize=(14,10))
df_close.plot(style=".")
plt.title("Scatter plot of closing")
plt.show()
```

```python
df["price_diff"]=df["Close"].shift(-1)-df["Close"]
df["price_diff"]
```

```python
df["daily_rn"]=df["price_diff"]/df["Close"]
df["daily_rn"]
```

```python
df["ra50"]=df["Close"].rolling(50).mean()
plt.figure(figsize=(14,10))
df["Close"].plot()
df["ra50"].plot()
plt.legend()
plt.show()
```

```python
df["ra10"]=df["Close"].rolling(10).mean()
```

```python
df=df.dropna()
df
```

```python
df["Shares"]=[1 if df.loc[ei,"ra10"]> df.loc[ei,"ra50"] else 0 for ei in df.index]
```

```python
df["Shares"]
```

```python
df["Close1"]=df["Close"].shift(-1)
df["profit"]=[df.loc[ei,"Close1"] - df.loc[ei,"Close"] if df.loc[ei,"Shares"]==1 else 0 for ei in df.index]
df["profit"].plot()
plt.axhline(y=0,color="red")
```
